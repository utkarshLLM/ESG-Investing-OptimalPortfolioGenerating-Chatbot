```python
import pandas as pd
import numpy as np
```

```python
string = 'esg_'
csv_extension = '.csv'
master_df = pd.DataFrame()
for i in range(2013, 2024):
    name = string + str(i)
    filename = name + csv_extension
    df = pd.read_csv(filename, index_col=0)
    master_df = pd.concat([master_df, df], axis=0)
```

```python
master_df
```

```python
master_df.to_csv('esg_2013to2023.csv', index=True)
```

#### Mapping

```python
msci_esg_legend = {'CCC':1, 'B':2, 'BB':3, 'BBB':4, 'A':5, 'AA':6, 'AAA':7}
```

#### Numeric ESG file

```python
master_df_numeric = master_df.replace(msci_esg_legend)
```

```python
master_df_numeric
```

```python
master_df_numeric.to_csv('esg_2013to2023_numeric.csv', index=True)
```
